```python
import focus_group
from focus_group import FocusGroup, Posture, Status, ColumnNames

diastolic, systolic, posture, status, predicted = focus_group.get_column_names()


focus_df = FocusGroup().df
standing_df = FocusGroup(poster_filter=Posture.standing).df
sitting_df = FocusGroup(poster_filter=Posture.sitting).df
normal_df = FocusGroup(status_filter=Status.normal).df
anxious_df = FocusGroup(status_filter=Status.anxious).df

normal_standing_df = FocusGroup(status_filter=Status.normal, poster_filter=Posture.standing).df
anxious_sitting_df = FocusGroup(status_filter=Status.anxious, poster_filter=Posture.sitting).df
normal_sitting_df = FocusGroup(status_filter=Status.normal, poster_filter=Posture.sitting).df
anxious_standing_df = FocusGroup(status_filter=Status.anxious, poster_filter=Posture.standing).df

print(anxious_standing_df.shape)
print(focus_df[systolic][0])
```

```python
import matplotlib.pylab as plt

plt.plot(normal_standing_df[diastolic])
plt.plot(normal_sitting_df[diastolic])
plt.plot(anxious_standing_df[diastolic])
plt.plot(anxious_sitting_df[diastolic])
plt.show()

plt.plot(anxious_df[diastolic])
plt.plot(normal_df[diastolic])
plt.show()

plt.plot(sitting_df[systolic])
plt.plot(standing_df[systolic])
plt.show()

plt.plot(anxious_df[systolic])
plt.plot(normal_df[systolic])
plt.show()

plt.plot(sitting_df[diastolic])
plt.plot(standing_df[diastolic])
plt.show()

plt.plot(anxious_df[diastolic])
plt.plot(normal_df[diastolic])
plt.show()

plt.plot(sitting_df[predicted])
plt.plot(standing_df[predicted])
plt.show()

plt.plot(anxious_df[predicted])
plt.plot(normal_df[predicted])
plt.show()


```
